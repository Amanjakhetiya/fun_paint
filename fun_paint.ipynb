```python
import cv2
import numpy as np
from collections import deque
```

```python
def p(text,image):
    cv2.imshow(text,image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
blueL=np.array([100,60,60])
blueU=np.array([120,255,255])

kernel=np.ones((5,5),np.uint8)

bindex=0
gindex=0
rindex=0
yindex=0

bpoint=[deque(maxlen=512)]
gpoint=[deque(maxlen=512)]
rpoint=[deque(maxlen=512)]
ypoint=[deque(maxlen=512)]

colorindex=0
colors=[(255,0,0),(0,255,0),(0,0,255),(0,255,255)]

paint=np.zeros((471,636,3))+255

paint=cv2.rectangle(paint,(40,65),(140,130),(0,0,0),2)
paint=cv2.rectangle(paint,(160,65),(255,130),colors[0],-1)
paint=cv2.rectangle(paint,(275,65),(370,130),colors[1],-1)
paint=cv2.rectangle(paint,(390,65),(485,130),colors[2],-1)
paint=cv2.rectangle(paint,(505,65),(600,130),colors[3],-1)

cv2.putText(paint,"CLEAR ALL",(49,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),2,cv2.LINE_AA)
cv2.putText(paint,"BLUE",(180,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
cv2.putText(paint,"GREEN",(293,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
cv2.putText(paint,"RED",(415,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
cv2.putText(paint,"YELLOW",(520,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(150,150,150),2,cv2.LINE_AA)
```

```python
cap=cv2.VideoCapture(0)

while 1:
    
    _,frame=cap.read()
    frame=cv2.flip(frame,1)
    hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    

    frame=cv2.rectangle(frame,(40,65),(140,130),(122,122,122),-1)
    frame=cv2.rectangle(frame,(160,65),(255,130),colors[0],-1)
    frame=cv2.rectangle(frame,(275,65),(370,130),colors[1],-1)
    frame=cv2.rectangle(frame,(390,65),(485,130),colors[2],-1)
    frame=cv2.rectangle(frame,(505,65),(600,130),colors[3],-1)

    cv2.putText(frame,"CLEAR ALL",(49,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,0),2,cv2.LINE_AA)
    cv2.putText(frame,"BLUE",(180,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"GREEN",(293,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"RED",(415,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2,cv2.LINE_AA)
    cv2.putText(frame,"YELLOW",(520,98),cv2.FONT_HERSHEY_SIMPLEX,0.5,(150,150,150),2,cv2.LINE_AA)
    
    
    
    mask=cv2.inRange(hsv,blueL,blueU)
    mask=cv2.erode(mask,kernel,iterations=2)
    mask=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
    mask=cv2.dilate(mask,kernel,iterations=1)
                
    cnts,hr=cv2.findContours(mask.copy(),cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
                
    center=None
                
    if len(cnts)>0:
                
        cnt=sorted(cnts,key=cv2.contourArea,reverse=True)[0]
        ((x,y),radius)=cv2.minEnclosingCircle(cnt)
        cv2.circle(frame,(int(x),int(y)),int(radius),(0,255,255),2)
        M=cv2.moments(cnt)
        center=(int(M['m10']/M['m00']),int(M['m01']/M['m00']))
                
        if center[1]<=130:
                
            if 140>=center[0]>=40:
                bpoint=[deque(maxlen=512)]
                gpoint=[deque(maxlen=512)]
                rpoint=[deque(maxlen=512)]
                ypoint=[deque(maxlen=512)]
                
                bindex=0
                gindex=0
                rindex=0
                yindex=0
                
                paint[67:,:,:]=255
                
            elif 255>=center[0]>=160:
                colorindex=0
            elif 370>=center[0]>=275:
                colorindex=1
            elif 485>=center[0]>=390:
                colorindex=2
            elif 600>=center[0]>=505:
                colorindex=3
        else:
                if colorindex==0:
                    bpoint[bindex].appendleft(center)
                elif colorindex==1:
                    gpoint[gindex].appendleft(center)
                elif colorindex==2:
                    rpoint[rindex].appendleft(center)
                elif colorindex==3:
                    ypoint[yindex].appendleft(center)
                
                
    else:
                bpoint.append(deque(maxlen=512))
                bindex=bindex+1
                gpoint.append(deque(maxlen=512))
                gindex=gindex+1
                rpoint.append(deque(maxlen=512))
                rindex=rindex+1
                ypoint.append(deque(maxlen=512))
                yindex=yindex+1
                
                
    points=[bpoint,gpoint,rpoint,ypoint]
    
    for i in range(len(points)):
        for j in range(len(points[i])):
            for k in range(1,len(points[i][j])):
                if points[i][j][k-1]==None or points[i][j][k]==None:
                      continue
                cv2.line(frame,points[i][j][k-1],points[i][j][k],colors[i],2)
                cv2.line(paint,points[i][j][k-1],points[i][j][k],colors[i],2)
                
                
                    
                
                
            
                
                
        
    
    
    
    cv2.imshow("final",frame)
    cv2.imshow("paint",paint)
    k=cv2.waitKey(1)
    if k==27:
        break
    
cap.release()
cv2.destroyAllWindows()
    
```
